### Ganadores: Cociente, Medio Cociente y Residuo

Los ganadores se definen de la siguiente forma:
1. Cociente: Cantidad de votos válidos entre la cantidad de curules (por partido)
2. Medio Cociente: Cociente entre 2. No pueden ganar los partidos que ya habían ganado por cociente
3. Residuo: Al partido del candidato que más votos ha tenido, contando todos los partidos afiliados

En el circuito 8-9 se tienen 3 curules a la fecha. En las votaciones de los 3 años se han mantenido de esa forma. Primero veremos los ganadores de cociente y medio cociente

```python
def dip_win(df): #Función para hacer el ranking por año
  df=df.groupby(['AÑO','Partido','Candidato'])[['votos']].sum()
  df.reset_index(inplace=True)
  years=df['AÑO'].unique()
  partido = df['Partido'].unique()
  df1= df.head(0)
  for i in years:
    for x in partido:
      df['diputado']=df[(df['AÑO'] == i)&(df['Partido'] == x)]['votos'].rank(ascending=False)
      df1=df1.append(df)

  df1.dropna(inplace=True)
  df1=df1[df1['diputado']==1]
  return df1  
```
